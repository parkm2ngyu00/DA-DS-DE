```python
import numpy as np
import pandas as pd

train = pd.read_csv('../data/wind_dataset/train.csv')
test = pd.read_csv('../data/wind_dataset/test.csv')

train
```

```python
import random, os

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(14) # Seed 고정
```

```python
train_x = train.drop(['ID', '풍속 (m/s)'], axis=1)
train_y = train['풍속 (m/s)']
```

```python
test_x = test.drop(columns=['ID'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le = le.fit(train_x['측정 시간대'])
train_x['측정 시간대'] = le.transform(train_x['측정 시간대'])
test_x['측정 시간대'] = le.transform(test_x['측정 시간대'])
```

```python
train_x.head(3)
```

```python
train_x.drop(['절대 온도(K)'], axis=1, inplace=True)
test_x.drop(['절대 온도(K)'], axis=1, inplace=True)
```

```python
train_x.head(3)
```

```python
train_y
```

```python
from sklearn.preprocessing import StandardScaler

std_scale = StandardScaler()
std_scale.fit(train_x)

X_train_std = std_scale.transform(train_x)
X_test_std = std_scale.transform(test_x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train_std, train_y, test_size=0.3)
```

### LinearRegression

```python
from sklearn.linear_model import LinearRegression

clf_lr = LinearRegression()
clf_lr.fit(X_train, y_train)
lr_pred = clf_lr.predict(X_test)
```

```python
print(clf_lr.coef_) # 가중치(w)
print(clf_lr.intercept_) # 바이어스(b)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, lr_pred)
print('선형 회귀 MAE:', mae)
```

### 릿지 & 라쏘 회귀

```python
from sklearn.linear_model import Ridge, Lasso

ridge = Ridge()
lasso = Lasso()

ridge.fit(X_test, y_test)
ridge_pred = ridge.predict(X_test)
lasso.fit(X_test, y_test)
lasso_pred = lasso.predict(X_test)

mae_ridge = mean_absolute_error(y_test, ridge_pred)
mae_lasso = mean_absolute_error(y_test, lasso_pred)

print('선형 회귀 MAE:', mae)
print('릿지 회귀 MAE:', mae_ridge)
print('라쏘 회귀 MAE:', mae_lasso)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(criterion='absolute_error')
dtr.fit(X_test, y_test)
dtr_pred = dtr.predict(X_test)

dtr_mae = mean_absolute_error(y_test, dtr_pred)
print(dtr_mae)
```
