```python
import pandas as pd

df = pd.read_csv('../data/wine_data.csv')
df
```

```python
from my_functions import create_feature_target, clf_eval

X_train, X_test, y_train, y_test = create_feature_target(df)
```

```python
from sklearn import svm
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV

param_grid = {
    'kernel' : ['linear', 'rbf'],
    'C' : [0.5, 1, 10, 100]
}
# kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
svc = svm.SVC(random_state=0)
grid_cv = GridSearchCV(svc, param_grid=param_grid, cv=5, scoring='accuracy')
grid_cv.fit(X_train, y_train)
```

```python
grid_cv.cv_results_
```

```python
import numpy as np

np.transpose(pd.DataFrame(grid_cv.cv_results_))
```

```python
grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
grid_cv.best_estimator_
```

```python
best_model = grid_cv.best_estimator_
best_model.fit(X_train, y_train)

# 재학습된 모델로 예측 수행
y_pred = best_model.predict(X_test)

from my_functions import clf_eval
clf_eval(y_test, y_pred)
```
