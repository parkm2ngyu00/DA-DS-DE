```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100,
                  n_features=2,
                  centers=5,
                  random_state=10)
```

```python
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1],
            c='gray',
            edgecolors='black',
            marker='o')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmc = KMeans(n_clusters=5,
             init='random',
             n_init=10,
             max_iter=100,
             random_state=0)

kmc.fit(X)
lable_kmc = kmc.labels_
print(lable_kmc)
```

```python
import pandas as pd

kmc_columns = ['kmc_comp1', 'kmc_comp2']
X_kmc_df = pd.DataFrame(X, columns=kmc_columns)
X_kmc_df['target'] = y
X_kmc_df['lable_kmc'] = lable_kmc
X_kmc_df.head(5)
```

```python
print(set(X_kmc_df['target']))
print(set(X_kmc_df['lable_kmc']))
```

```python
df = X_kmc_df
markers = ['o', 'x', '^', 's', '*']

for i, mark in enumerate(markers):
    df_i = df[df['lable_kmc']==i]
    target_i = i
    X1 = df_i['kmc_comp1']
    X2 = df_i['kmc_comp2']
    plt.scatter(X1, X2, marker=mark, label=target_i)

plt.xlabel('kmc_component1')
plt.ylabel('kmc_component2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(X, lable_kmc)
print(sil_score)
```
