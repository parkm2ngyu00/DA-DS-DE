```python
import pandas as pd

df = pd.read_csv('../data/wine_data.csv')
df
```

```python
from my_functions import create_feature_target, clf_eval

X_train, X_test, y_train, y_test = create_feature_target(df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
print(X_train_pca.shape)
print(X_test_pca.shape)
```

```python
print(pca.singular_values_)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

```python
print(pca.explained_variance_ratio_)
```

```python
pca_columns = ['pca_comp1', 'pca_comp2']
X_train_pca_df = pd.DataFrame(X_train_pca, columns=pca_columns)
y_tn1 = y_test.reset_index(drop=True)
X_train_pca_df['target'] = y_tn1
X_train_pca_df.head(10)
```
