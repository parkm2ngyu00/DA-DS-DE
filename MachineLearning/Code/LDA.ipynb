```python
import pandas as pd

df = pd.read_csv('../data/wine_data.csv')
df
```

```python
from my_functions import create_feature_target, clf_eval

X_train, X_test, y_train, y_test = create_feature_target(df)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
X_tn_lda = lda.transform(X_train)
X_te_lda = lda.transform(X_test)
```

```python
print(lda.intercept_)
```

```python
print(lda.coef_)
```

```python
lda_columns = ['lda_comp1', 'lda_comp2']
X_tn_lda_df = pd.DataFrame(X_tn_lda, columns=lda_columns)
X_tn_lda_df['target'] = y_train.reset_index(drop=True)
X_tn_lda_df.head(5)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf_rf_lda = RandomForestClassifier(max_depth=2, random_state=0)
clf_rf_lda.fit(X_tn_lda, y_train)
pred_rf_lda = clf_rf_lda.predict(X_te_lda)
```

```python
clf_eval(y_test, pred_rf_lda)
```
