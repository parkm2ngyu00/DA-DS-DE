```python
# 라이브러리를 임포트
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 흑백 이미지로 로드
image = cv2.imread("cat&dog.jpg", 
                   cv2.IMREAD_GRAYSCALE)
```

```python
# 이미지를 출력
plt.imshow(image, cmap="gray"), plt.axis("off")
plt.show()
```

## 2-이미지 크기 변경하기

```python
# 흑백 이미지로 로드
image = cv2.imread("cat&dog.jpg", cv2.IMREAD_GRAYSCALE)

# 이미지 크기를 50x50 픽셀로 변경
image_50x50 = cv2.resize(image, (50, 50))

# 이미지를 출력
plt.imshow(image_50x50, cmap="gray"), plt.axis("off")
plt.show()
```

## 3-머신러닝 특성 만들기

```python
# 흑백 이미지로 로드
image = cv2.imread("cat&dog.jpg", cv2.IMREAD_GRAYSCALE)

# 이미지를 10x10 픽셀 크기로 변환
image_10x10 = cv2.resize(image, (10, 10))

# 이미지 데이터를 1차원 벡터로 변환
image_10x10.flatten()
```

```python
plt.imshow(image_10x10, cmap="gray"), plt.axis("off")
plt.show()
```

```python
image_10x10.shape
```

```python
image_10x10.flatten().shape
```

```python
# 컬러 이미지로 로드
image_color = cv2.imread("cat&dog.jpg", cv2.IMREAD_COLOR)

# 이미지를 10 × 10 픽셀 크기로 변환
image_color_10x10 = cv2.resize(image_color, (10, 10))

# 이미지 데이터를 1차원 벡터로 변환하고 차원을 출력
image_color_10x10.flatten().shape
```

```python
# 이미지를 출력
plt.imshow(image_color_10x10), plt.axis("off")
plt.show()
```
