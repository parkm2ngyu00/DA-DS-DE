### 2. 와인 추천하기

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('../Data/wine_data.csv')

df.shape
```

```python
df.head(3)
```

```python
df = df[['title', 'description']].iloc[:5000]
```

```python
df.head(10)
```

```python
df['description'].isnull().values.any()    # 결손 데이터 없는 것 확인
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english') # 크게 의미없는 단어들을 제외해준다.
tfidf_matrix = tfidf.fit_transform(df['description'])
tfidf_matrix.shape
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)     # 코사인 유사도 계산
cosine_sim.shape
```

```python
cosine_sim
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()    # 중복 제거
indices
```

```python
def recommend(title):
    idx = indices[title]
    sim_list = list(enumerate(cosine_sim[idx]))
    sim_list = sorted(sim_list, key=lambda x : x[1], reverse=True)
    top10 = sim_list[1:11]
    top10_list = [x[0] for x in top10]
    return df.iloc[top10_list].title
```

```python
recommend('Quinta dos Avidagos 2011 Avidagos Red (Douro)')
```
