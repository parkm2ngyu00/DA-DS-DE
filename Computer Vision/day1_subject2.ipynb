```python
image_path = './img/image3.jpg'
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#  .
image = cv2.imread(image_path)

# BGR  .
b, g, r = cv2.split(image)

#     .
equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)

#    .
equalized_image = cv2.merge((equalized_b, equalized_g, equalized_r))

#   .
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.title('RGB Equalized Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#  .
image = cv2.imread(image_path)

#  HSV   .
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# H, S, V  .
h, s, v = cv2.split(hsv_image)

#  .
h_hist = cv2.calcHist([h], [0], None, [256], [0, 256])
s_hist = cv2.calcHist([s], [0], None, [256], [0, 256])
v_hist = cv2.calcHist([v], [0], None, [256], [0, 256])

#   .
equalized_h = cv2.equalizeHist(h)
equalized_s = cv2.equalizeHist(s)
equalized_v = cv2.equalizeHist(v)

#  .
h_equalized_hist = cv2.calcHist([equalized_h], [0], None, [256], [0, 256])
s_equalized_hist = cv2.calcHist([equalized_s], [0], None, [256], [0, 256])
v_equalized_hist = cv2.calcHist([equalized_v], [0], None, [256], [0, 256])

#  H, S, V   .
equalized_hsv_image = cv2.merge((equalized_h, equalized_s, equalized_v))

# HSV  BGR   .
equalized_image = cv2.cvtColor(equalized_hsv_image, cv2.COLOR_HSV2BGR)

#   .
plt.figure(figsize=(15, 10))

plt.subplot(3, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(3, 3, 2)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.title('HSV Equalized')
plt.axis('off')

plt.subplot(3, 3, 4)
plt.plot(h_hist, color='r')
plt.title('H Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(3, 3, 5)
plt.plot(s_hist, color='g')
plt.title('S Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(3, 3, 6)
plt.plot(v_hist, color='b')
plt.title('V Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(3, 3, 7)
plt.plot(h_equalized_hist, color='r')
plt.title('H Equalized Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(3, 3, 8)
plt.plot(s_equalized_hist, color='g')
plt.title('S Equalized Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(3, 3, 9)
plt.plot(v_equalized_hist, color='b')
plt.title('V Equalized Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#  .
image = cv2.imread(image_path)

#  HSV   .
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# H, S, V  .
h, s, v = cv2.split(hsv_image)

#  .
h_hist = cv2.calcHist([h], [0], None, [256], [0, 256])
s_hist = cv2.calcHist([s], [0], None, [256], [0, 256])
v_hist = cv2.calcHist([v], [0], None, [256], [0, 256])

#   .
equalized_h = cv2.equalizeHist(h)
equalized_s = cv2.equalizeHist(s)
equalized_v = cv2.equalizeHist(v)

# H, S, V      CDF 
equalized_h_hist = cv2.calcHist([equalized_h], [0], None, [256], [0, 256])
equalized_s_hist = cv2.calcHist([equalized_s], [0], None, [256], [0, 256])
equalized_v_hist = cv2.calcHist([equalized_v], [0], None, [256], [0, 256])

equalized_h_cdf = np.cumsum(equalized_h_hist)
equalized_s_cdf = np.cumsum(equalized_s_hist)
equalized_v_cdf = np.cumsum(equalized_v_hist)

equalized_h_cdf_normalized = equalized_h_cdf / equalized_h_cdf[-1]
equalized_s_cdf_normalized = equalized_s_cdf / equalized_s_cdf[-1]
equalized_v_cdf_normalized = equalized_v_cdf / equalized_v_cdf[-1]

#  CDF 
plt.figure(figsize=(15, 20))

plt.subplot(4, 3, 1)
plt.plot(h_cdf_normalized)
plt.title('Original H CDF')
plt.xlabel('Pixel Value')
plt.ylabel('CDF')

plt.subplot(4, 3, 2)
plt.plot(s_cdf_normalized)
plt.title('Original S CDF')
plt.xlabel('Pixel Value')
plt.ylabel('CDF')

plt.subplot(4, 3, 3)
plt.plot(v_cdf_normalized)
plt.title('Original V CDF')
plt.xlabel('Pixel Value')
plt.ylabel('CDF')

plt.subplot(4, 3, 4)
plt.plot(equalized_h_cdf_normalized)
plt.title('Equalized H CDF')
plt.xlabel('Pixel Value')
plt.ylabel('CDF')

plt.subplot(4, 3, 5)
plt.plot(equalized_s_cdf_normalized)
plt.title('Equalized S CDF')
plt.xlabel('Pixel Value')
plt.ylabel('CDF')

plt.subplot(4, 3, 6)
plt.plot(equalized_v_cdf_normalized)
plt.title('Equalized V CDF')
plt.xlabel('Pixel Value')
plt.ylabel('CDF')

plt.tight_layout()
plt.show()
```
